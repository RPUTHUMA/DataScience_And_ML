### converts categorical column to numeric features


```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
## Create a simple dataframe 
df = pd.DataFrame({
    'color': ['red', 'blue', 'green', 'green', 'red', 'blue']
})
```

```python
df['color'].shape
```

```python
df
```

```python
df[['color']].shape
```

```python
encoder=OneHotEncoder()
encoded =encoder.fit_transform(df[['color']]).toarray() ## arranged alphabetically
```

```python
encoded1 =encoder.fit_transform(df[['color']]) ## arranged alphabetically
encoded1
```

```python
?encoder.fit_transform
```

```python
encoded 
```

```python
encoded_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
encoded_df
```

```python
#for new data
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df,encoded_df],axis=1)
```
